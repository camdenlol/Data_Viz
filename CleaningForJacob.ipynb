# Scrapping NFL Player Data

I found this site with the nfl players by position with the data we were interested in.  Here is an attempt to scrape it!

```python
import requests
import pandas as pa
from bs4 import BeautifulSoup


r = requests.get('https://newsday.sportsdirectinc.com/football/nfl-players.aspx?page=/data/nfl/players/A_players.html')
html_contents = r.text
html_soup = BeautifulSoup(html_contents,"lxml")
```

```python

tables = html_soup.find_all('table')
len(tables)
```

```python
df = pa.read_html(str(tables))[2]
df.columns = df.iloc[0,:]
df = df.iloc[1:,:]
df
```

```python
import string
let = string.ascii_uppercase
let[1:]
```

```python
for i in let[1:]:
  r = requests.get('https://newsday.sportsdirectinc.com/football/nfl-players.aspx?page=/data/nfl/players/' + i +'_players.html')
  html_contents = r.text
  html_soup = BeautifulSoup(html_contents,"lxml")
  tables = html_soup.find_all('table')
  dftemp = pa.read_html(str(tables))[2]
  dftemp.columns = dftemp.iloc[0,:]
  dftemp = dftemp.iloc[1:,:]
  df = pa.concat([df,dftemp])

df
```

```python
df.to_csv("players.csv", index = False)
```
