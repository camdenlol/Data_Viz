# Is There a Correlation Between the Detroit Lion's Record and Homicide in Detroit?

Using the Federal Bureau of Investigation's Crime Data Explorer and Wikipedia, we can create visualizations showing the record of the Detroit Lions and the homicide rates in Detroit to see if there is any correlation.

```python
import pandas as pa
import matplotlib.pyplot as plt
```

```python
years = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
homicides = [355, 297, 279, 273, 290, 245, 243, 281, 307, 306, 293]


plt.figure(figsize=(10, 6))
plt.plot(years, homicides, marker='o', linestyle='-')
plt.title('Homicides Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Homicides')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
wins = [25, 43.8, 68.8, 43.8, 56.3, 56.3, 37.5, 21.9, 31.3, 20.6, 52.9]

plt.figure(figsize=(10, 6))
plt.plot(years, wins, marker='o', linestyle='-')
plt.title('Lions Record Over the Years')
plt.xlabel('Year')
plt.ylabel('Winning %')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

What we are looking for in these two graphs is when the win percentage is high, and the homicide rate is low. This is really interesting. I believe that we can make a solid argument for the correlation between football and homicides in Detroit. The FBI doesn't have data for 2023, but that was a good year for the Lions with a 70.6% win ratio. After a quick Google search, multiple news sources state that homicides in Detroit are at their lowest since 1966. We also see from the graphs above that when the record improved in 2013-2016, the homicide rate went down, and it shot up when the record declined in 2019-2021. Of course, there are other factors such as COVID-19, elections, gun control, etc., happening, but the data shows that this could play a role.
